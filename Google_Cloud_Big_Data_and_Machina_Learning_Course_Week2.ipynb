ideo#3 cont'd:
 - Apache Beam is a portable open source data processing programming model
 - Can be ran is a highly distributed fashion (distributed processing)
 - It is unified: meaning a single programming model can be used for both batch and streaming use cases
 - It is portable: can work in multiple execution environments so you are not locked in to a vendor (Cloud Dataflow, Apach Spark etc.)
 - Open Source and Extensible: You can write and share new SDKs, IO Connectors and transformation libraries
 - Apache Beam libraries are written in Python, Java and Go
 - Apache Beam create a model representation of your code which is portable across many runners (runners pass off yout model to an execution environment whcih you could run in many different possible engines such as Cloud Dataflow)
